### Question 1: Use the yfinance to Extract Stock Data

Using the `Ticker` function, enter the ticker symbol of the stock we want to extract data on to create a ticker object. The stock is Tesla and its ticker symbol is `TSLA`.

Using the ticker object and the function `history`, extract stock information and save it in a dataframe named `tesla_stock`. Set the `period` parameter to `max` so we get information for the maximum amount of time.

Reset the index, save, and display the first five rows of the `tesla_stock` dataframe using the `head` function. Take a screenshot of the results.


```python
import yfinance as yf

# Descargar datos de Tesla
tesla_data = yf.Ticker("TSLA")
tesla_stock = tesla_data.history(period="max")

# Restablecer el índice y mostrar las primeras filas
tesla_stock.reset_index(inplace=True)
tesla_stock.head()
```

### Question 6: Create GameStop Dashboard

Create a dashboard to display GameStop's stock price and revenue data. Use matplotlib to create two subplots:
1. A line chart showing GameStop's stock price over time.
2. A bar chart showing GameStop's revenue over time.


```python
import matplotlib.pyplot as plt
import yfinance as yf

# Función para dibujar el gráfico
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], color='blue', label='Close Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price")
    plt.legend()
    plt.grid(True)
    plt.show()

# Descargar los datos de GameStop
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Restablecer el índice para usar 'Date' como columna
gme_data.reset_index(inplace=True)

# Llamar a la función para graficar
make_graph(gme_data, "GameStop Stock Price Over Time")
```
